Initial exploratory analysis of the Honolulu Traffic Incident data found here: https://data.honolulu.gov/Public-Safety/Traffic-Incidents/ykb6-n5th/about_data

Grab database url and app token information from .env file

```python
%load_ext dotenv
%dotenv
import os

database_url = os.getenv("DATABASE_URL")
app_token = os.getenv("APP_TOKEN")
```

Pull Honolulu Traffic Incident data from Socratic API.

Dataset details: Traffic Incidents in the last 24 hours updated every 5 minutes.

```python
import pandas as pd
from sodapy import Socrata

# Unauthenticated client only works with public data sets. Note 'None'
# in place of application token, and no username or password:
client = Socrata("data.honolulu.gov", app_token)

# First 2000 results, returned as JSON from API / converted to Python list of
# dictionaries by sodapy.
results = client.get("ykb6-n5th", limit=2000)

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)

print(results_df)
```
